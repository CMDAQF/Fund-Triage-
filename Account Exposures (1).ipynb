```python
import pandas as pd

yrdata = pd.read_csv(r'C:\your file path\Tickers.csv')
acct = pd.read_csv(r'C:\your file path \Positions.csv')

def ycount_occurrences(yrdata, acct, columns1, columns2):
    count_dict_yellow = {}
    count_dict_red = {}

    for column in columns1:
        if column == "Yellow":
            count_dict = count_dict_yellow
        elif column == "Red":
            count_dict = count_dict_red

        for item in yrdata[column]:
            count_dict[item] = 0

    for column in columns2:
        for item in acct[column].dropna(): #removes NaN values from counting
            if item in count_dict_yellow:
                count_dict_yellow[item] += 1
            elif item in count_dict_red:
                count_dict_red[item] += 1

    return count_dict_yellow, count_dict_red

columns1 = ["Yellow", "Red"]  
columns2 = acct.columns.tolist()  # Get the list of column names in acct

output_data = []

for column in columns2:
    yellow_occurrences, red_occurrences = ycount_occurrences(yrdata, acct, columns1, [column])

    total_yellow_occurrences = sum(yellow_occurrences.values())
    total_red_occurrences = sum(red_occurrences.values())

    if total_yellow_occurrences > 0:
        yellow_items = {item: count for item, count in yellow_occurrences.items() if count > 0}
    else:
        yellow_items = {}

    if total_red_occurrences > 0:
        red_items = {item: count for item, count in red_occurrences.items() if count > 0}
    else:
        red_items = {}

    Exposure = (total_red_occurrences * 2) + total_yellow_occurrences

    output_data.append({"Column": column,
                        "Yellow Occurrences": yellow_items,
                        "Red Occurrences": red_items,
                        "Total Yellow Occurrences": total_yellow_occurrences,
                        "Total Red Occurrences": total_red_occurrences,
                        "Account exposure is": Exposure})

output_df = pd.DataFrame(output_data)

output_file = "occurrences_output.xlsx"
output_df.to_excel(output_file, index=False)

print("Output saved to:", output_file)
```
